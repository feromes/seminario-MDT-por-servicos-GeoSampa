## Definindo uma região de estudo

> Parque Estadual do Jaraguá

```python
from owslib.csw import CatalogueServiceWeb
datageo = WebMapService('http://datageo.ambiente.sp.gov.br/geoserver/datageowms/ows?SERVICE=WMS', version='1.3.0')
```

```python
print(datageo.contents['UCS_ESTADUAIS_PI_SP_POL'].title)
```

```python
# Define uma coordenada para captura da feição, UTM SIRGAS 2000
c = (319526,7404078)

pe_jaragua = datageo.getfeatureinfo(layers = ['UCS_ESTADUAIS_PI_SP_POL'],
                          format = 'image/png',
                          info_format = 'application/json',
                          srs = 'EPSG:31983',
                          bbox = (c[0] - 1, c[1] - 1, c[0] + 1, c[1] + 1),
                          size = (2, 2), 
                          xy = (1, 1))
```

```python
import json
pe_jaragua_json = json.load(pe_jaragua)
```

```python
pe_jaragua_json['features'][0]['properties']
```

```python
from shapely.geometry import Polygon
polygon = Polygon(pe_jaragua_json['features'][0]['geometry']['coordinates'][0][0])
```

```python
print(f"A área é de {polygon.area}m² e o perímetro é de {polygon.length}m")
```

```python
polygon
```

## Obtendo os pontos do terreno, LiDAR MDS 3D

* Nuvem de pontos com mais de 31 bilhões de pontos, aja memória
* Mais de 5500 folhas SCM, quadriculas com 19" de lado para download
* Solução encontrada: EPT (Entwine Point Tiles)
* Serviço hospedado em nuvem Amazon OpenData

```python
from IPython.display import IFrame
IFrame(src="https://potree.entwine.io/", width='100%', height='500px')
```

```python
IFrame(src="https://visualizador-laz-web.s3-sa-east-1.amazonaws.com/index.html?&mdt=1&xmax=320918&ymax=7405811&xmin=318304&ymin=7402572", width='100%', height='500px')
```

```python
import pdal

bb = polygon.bounds
cropper = [{
      "type": "readers.ept",
      "filename": "https://ept-m3dc-pmsp.s3-sa-east-1.amazonaws.com/ept.json",
      "bounds": f"([{bb[0]}, {bb[2]}],[{bb[1]}, {bb[3]}])",
      "resolution": 5,
      "threads":6
    },{
      "type":"filters.range",
      "limits":"Classification[2:2]"
    },{
      "filename":"./resultados/dtm_mdt.tif",
      "gdaldriver":"GTiff",
      "output_type":"all",
      "resolution":"5.0",
      "type": "writers.gdal"}]

pipeline = pdal.Pipeline(json.dumps(cropper))
pipeline.validate()
%time n_points = pipeline.execute()
print(f'O serviço selecionou {n_points} pontos em uma área de {polygon.area} m², uma densidade de {n_points/polygon.area}')
```

```python
import pandas as pd
```

```python
arr = pipeline.arrays[0]
df = pd.DataFrame(arr)
df.head(10)
```

## Demonstrando e exportando os resultados

```python
import rasterio
from rasterio import fill
from rasterio.plot import show, show_hist
import numpy as np
```

```python
raster = rasterio.open('./resultados/dtm_mdt.tif')
```

```python
raster_filled = rasterio.fill.fillnodata(raster.read(3), mask=raster.read_masks(3), max_search_distance=10)
```

```python
raster_filled = np.ma.masked_array(raster_filled, mask=(raster_filled == -9999.))
```

```python
import matplotlib.pylab as plt
plt.rcParams['figure.dpi'] = 200
```

```python
show(raster_filled, cmap='terrain')
```

```python
show(raster_filled, cmap='terrain', contour=True)
```

```python
from rasterio.plot import show_hist
```

```python
show_hist(raster_filled, bins=50, lw=0.0, stacked=True, alpha=1, histtype='stepfilled', title="Histogram")
```
